# Question 9 - Hypothesis Testing

```python
import numpy as np

class CorrelationPermute(thinkstats2.HypothesisTest):
    
    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat
    
    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys

steps = daily_merged['TotalSteps']
sleep = daily_merged['TotalMinutesAsleep']
data = daily_merged.TotalSteps.values, daily_merged.TotalMinutesAsleep.values
ht = CorrelationPermute(data)
pvalue = ht.PValue()
```
